## 高级选项

除了本地编辑外，还有两件重要的事情：在 markdown 格式中编辑笔记本以及远程运行 Jupyter。后者在我们希望在更快的服务器上运行代码时很重要。前者很重要，因为 Jupyter 的原生 ipynb 格式存储了许多与内容无关的辅助数据，主要是与代码如何及在哪里运行有关的数据。这会使 Git 感到困惑，使得审查贡献非常困难。幸运的是，有一种替代方案——直接在 Jupyter 中以 markdown 格式进行原生编辑。

### Jupyter 中的 Markdown 文件

如果您希望对本书的内容做出贡献，则需要修改 GitHub 上的源文件（md 文件，而不是 ipynb 文件）。使用 notedown 插件，我们可以在 Jupyter 中直接修改 md 格式的笔记本。


首先，安装 notedown 插件，运行 Jupyter Notebook，并加载插件：

```
pip install d2l-notedown  # 您可能需要卸载原始的 notedown。
jupyter notebook --NotebookApp.contents_manager_class='notedown.NotedownContentsManager'
```

您还可以在每次运行 Jupyter Notebook 时默认启用 notedown 插件。首先，生成 Jupyter Notebook 配置文件（如果已经生成，可以跳过此步骤）。

```
jupyter notebook --generate-config
```

然后，在 Jupyter Notebook 配置文件（对于 Linux 或 macOS，通常位于路径 `~/.jupyter/jupyter_notebook_config.py`）的末尾添加以下行：

```
c.NotebookApp.contents_manager_class = 'notedown.NotedownContentsManager'
```

之后，只需运行 `jupyter notebook` 命令即可默认启用 notedown 插件。

### 在远程服务器上运行 Jupyter 笔记本

有时，您可能希望在远程服务器上运行 Jupyter 笔记本并通过本地计算机上的浏览器访问它。如果您的本地机器安装了 Linux 或 macOS（Windows 也可以通过第三方软件如 PuTTY 支持此功能），则可以使用端口转发：

```
ssh myserver -L 8888:localhost:8888
```

上述字符串 `myserver` 是远程服务器的地址。然后我们可以使用 http://localhost:8888 访问运行 Jupyter 笔记本的远程服务器 `myserver`。稍后在本附录中，我们将详细介绍如何在 AWS 实例上运行 Jupyter 笔记本。

### 计时

我们可以使用 `ExecuteTime` 插件来计时 Jupyter 笔记本中每个代码单元格的执行时间。使用以下命令安装插件：

```
pip install jupyter_contrib_nbextensions
jupyter contrib nbextension install --user
jupyter nbextension enable execute_time/ExecuteTime
```

* 我们可以通过端口转发在远程服务器上运行 Jupyter 笔记本。
