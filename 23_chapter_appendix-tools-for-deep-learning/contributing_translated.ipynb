# 为本书做贡献
:label:`sec_how_to_contribute`

[读者](https://github.com/d2l-ai/d2l-en/graphs/contributors)的贡献帮助我们改进这本书。如果您发现错别字、过时的链接、我们认为遗漏了引用的地方、代码不够优雅或解释不清楚的地方，请回馈并帮助我们帮助其他读者。虽然传统书籍在印刷版本之间（因此在更正错别字之间）可能需要数年时间，但通常只需要几小时到几天就能将改进纳入本书。这都得益于版本控制和持续集成 (CI) 测试。要做到这一点，您需要向 GitHub 仓库提交一个 [pull request](https://github.com/d2l-ai/d2l-en/pulls)。当您的 pull request 被作者合并到代码仓库中时，您将成为一名贡献者。

## 提交小改动

最常见的贡献是编辑一句话或修复错别字。我们建议您找到 [GitHub 仓库](https://github.com/d2l-ai/d2l-en)中的源文件并直接编辑该文件。例如，您可以使用 [Find file](https://github.com/d2l-ai/d2l-en/find/master) 按钮 (:numref:`fig_edit_file`) 来搜索文件以定位源文件（一个 markdown 文件）。然后点击右上角的 "Edit this file" 按钮来修改 markdown 文件。

![在 Github 上编辑文件。](../img/edit-file.png)
:width:`300px`
:label:`fig_edit_file`

完成后，在页面底部的 "Propose file change" 面板中填写您的更改描述，然后点击 "Propose file change" 按钮。它会将您重定向到一个新页面来审核您的更改 (:numref:`fig_git_createpr`)。如果一切正常，您可以通过点击 "Create pull request" 按钮来提交 pull request。

## 提出重大更改

如果您计划更新大量文本或代码，则需要更多地了解本书使用的格式。源文件基于 [markdown 格式](https://daringfireball.net/projects/markdown/syntax)，并通过 [D2L-Book](http://book.d2l.ai/user/markdown.html) 包扩展了一些功能，如引用公式、图片、章节和引用。您可以使用任何 markdown 编辑器打开这些文件并进行修改。

如果您想更改代码，我们建议您按照 :numref:`sec_jupyter` 中的描述使用 Jupyter Notebook 打开这些 markdown 文件，以便您可以运行和测试您的更改。请记住在提交更改之前清除所有输出，因为我们的 CI 系统将执行您更新的部分以生成输出。

某些部分可能支持多种框架实现。
如果您添加了一个新的代码块，请使用 `%%tab` 在开始行标记此块。例如，
`%%tab pytorch` 用于 PyTorch 代码块，`%%tab tensorflow` 用于 TensorFlow 代码块，或 `%%tab all` 用于所有实现的共享代码块。您可以参考 `d2lbook` 包获取更多信息。

## 提交重大更改

我们建议您使用标准的 Git 过程来提交重大更改。简而言之，该过程如 :numref:`fig_contribute` 所示。

![为本书做贡献。](../img/contribute.svg)
:label:`fig_contribute`

我们将详细指导您完成这些步骤。如果您已经熟悉 Git，可以跳过本节。为了具体说明，我们假设贡献者的用户名是 "astonzhang"。

### 安装 Git

Git 开源书描述了 [如何安装 Git](https://git-scm.com/book/en/v2)。这通常通过在 Ubuntu Linux 上使用 `apt install git`，在 macOS 上安装 Xcode 开发工具，或使用 GitHub 的 [桌面客户端](https://desktop.github.com) 来完成。如果您没有 GitHub 账户，您需要注册一个。

### 登录 GitHub

在浏览器中输入本书代码仓库的 [地址](https://github.com/d2l-ai/d2l-en/)。点击 :numref:`fig_git_fork` 右上角红色框中的 `Fork` 按钮，制作本书仓库的副本。现在这是 *您的副本*，您可以按您想要的方式进行修改。

![代码仓库页面。](../img/git-fork.png)
:width:`700px`
:label:`fig_git_fork`


现在，本书的代码仓库将被 fork（即复制）到您的用户名下，例如 `astonzhang/d2l-en` 如 :numref:`fig_git_forked` 左上角所示。

![fork 后的代码仓库。](../img/git-forked.png)
:width:`700px`
:label:`fig_git_forked`

### 克隆仓库

要克隆仓库（即创建本地副本），我们需要获取其仓库地址。:numref:`fig_git_clone` 中的绿色按钮显示了这一点。确保如果您决定长期保留这个 fork，您的本地副本与主仓库保持同步。目前只需按照 :ref:`chap_installation` 中的说明开始即可。主要区别在于您现在下载的是 *您自己的 fork* 仓库。

![克隆仓库。](../img/git-clone.png)
:width:`700px`
:label:`fig_git_clone`

```
# 将 your_github_username 替换为您的 GitHub 用户名
git clone https://github.com/your_github_username/d2l-en.git
```

### 编辑和推送

现在是时候编辑这本书了。最好按照 :numref:`sec_jupyter` 中的说明在 Jupyter Notebook 中进行编辑。进行更改并检查它们是否正确。假设我们在文件 `~/d2l-en/chapter_appendix-tools-for-deep-learning/contributing.md` 中修改了一个错别字。
然后您可以检查哪些文件已被更改。

此时 Git 会提示 `chapter_appendix-tools-for-deep-learning/contributing.md` 文件已被修改。

```
mylaptop:d2l-en me$ git status
On branch master
Your branch is up-to-date with 'origin/master'.

Changes not staged for commit:
  (use "git add <file>..." to update what will be committed)
  (use "git checkout -- <file>..." to discard changes in working directory)

	modified:   chapter_appendix-tools-for-deep-learning/contributing.md
```

确认这是您想要的后，执行以下命令：

```
git add chapter_appendix-tools-for-deep-learning/contributing.md
git commit -m 'Fix a typo in git documentation'
git push
```

更改后的代码将位于您个人的仓库副本中。要请求添加您的更改，您必须为本书的官方仓库创建一个 pull request。

### 提交 Pull Requests

如 :numref:`fig_git_newpr` 所示，前往您在 GitHub 上的仓库副本并选择 "New pull request"。这将打开一个屏幕，显示您的编辑与本书主仓库当前内容之间的差异。

![新建 pull request。](../img/git-newpr.png)
:width:`700px`
:label:`fig_git_newpr`


最后，通过点击 :numref:`fig_git_createpr` 中所示的按钮来提交 pull request。请务必在 pull request 中描述您所做的更改。
这将使作者更容易审查并将更改合并到书中。根据更改的不同，可能会立即接受、拒绝，或者更有可能的是，您会收到一些关于更改的反馈。一旦您将其整合完毕，就可以继续进行了。

![创建 pull request。](../img/git-createpr.png)
:width:`700px`
:label:`fig_git_createpr`


## 总结

* 您可以使用 GitHub 为本书做贡献。
* 对于小改动，可以直接在 GitHub 上编辑文件。
* 对于重大更改，请先 fork 仓库，在本地进行编辑，并在准备好后再贡献回来。
* pull requests 是如何捆绑贡献的。尽量不要提交巨大的 pull requests，因为这会使它们难以理解和合并。最好发送几个较小的 pull requests。
